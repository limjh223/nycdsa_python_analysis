```python
#getting monthly average of max temp from july 2008 to oct 2021
max_avg = weather['maxtempF'].resample('M').mean()
max_std = weather['maxtempF'].resample('M').std()
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
plt.style.use('default')
max_avg.plot(label = 'Avg')
max_std.plot(label = 'Std Dev')
plt.title('Monthly High Temperature')
plt.ylabel('Temp (deg F)')
plt.legend()
plt.ylim(0, 110)
plt.savefig('Monthly High Temperature.png')
```

```python
#repeating the same process with the low temps
min_avg = weather['mintempF'].resample('M').mean()
min_std = weather['mintempF'].resample('M').std()
```

```python
min_avg.plot(c = 'k', 
             label = 'Avg')
min_std.plot(c = 'g', 
             label = 'std dev')
plt.ylabel('Temp (deg F)')
plt.style.use('default')
plt.legend()
plt.ylim(0, 100)
plt.title('Monthly Low Temperature')
plt.savefig('Monthly Low Temperature.png')
```

```python
#repeating the same process with precipitation

precip_avg = weather['totalprecipIn'].resample('M').mean()
precip_std = weather['totalprecipIn'].resample('M').std()
```

```python
precip_avg.plot(c='#ffa500', 
                label = 'Avg')
precip_std.plot(c='g',
                label = 'Std Dev')
plt.title('Monthly Precipitation')
plt.legend()
plt.ylabel('Precipitation (in)')
plt.ylim(0, 1.75)
plt.xlabel('Date')
plt.savefig('Monthly Precipitation')
```

```python

windsp_avg = weather['windspeedMiles'].resample('M').mean()


windsp_std = weather['windspeedMiles'].resample('M').std()

```

```python
import datetime as datetime 
plt.style.use('default')

windsp_avg = pd.DataFrame(windsp_avg)
dates = pd.date_range(start = datetime.date(2008,7,31), periods=160, freq='M')
plt.scatter(x = np.array(dates), 
            y = np.array(windsp_avg['windspeedMiles']), 
            marker='.', 
            c='k', 
            label = 'average')


windsp_std = pd.DataFrame(windsp_std)
plt.scatter(x = np.array(dates), 
            y = np.array(windsp_std['windspeedMiles']), 
            marker='.', 
            c='r',
            label = 'std dev')
plt.legend()
plt.ylabel('Wind Speed (MPH)')
plt.xlabel('Year')
plt.ylim(0, 15)
plt.title('Wind Speed Since 2008')
plt.savefig('Wind Speed Since 2008.png')
```

```python
# creating a column for the difference between max & min temperature
# creating another column for the difference between wind gust & speed

def sub(a,b):
    return a - b
weather['temp_range'] = weather.apply(lambda row: sub(row['maxtempF'], row['mintempF']), axis = 1)
col_temprange = weather.pop('temp_range')
weather.insert(2, 'temp_range', col_temprange)

weather['wind_range'] = weather.apply(lambda row: sub(row['WindGustMiles'], row['windspeedMiles']), axis = 1)
col_windrange = weather.pop('wind_range')
col_windgust = weather.pop('WindGustMiles')

weather.insert(5, 'WindGustMiles', col_windgust)
weather.insert(7, 'wind_range', col_windrange)
```

```python

per = weather.date.dt.to_period("M")
per2 = weather.date.dt.to_period("Y")
monthly_precip = weather.groupby(per)['totalprecipIn'].agg('sum')
yearly_precip = weather.groupby(per2)['totalprecipIn'].agg('sum')
plt.style.use('seaborn')
plt.title('Monthly and Yearly Total Precipitation')
monthly_precip.plot(label = 'Monthly Total')
yearly_precip.plot(label = 'Yearly Total')
plt.xlim('2010', '2020')
plt.ylim(0, 100)
plt.xlabel('Year', c='k')
plt.ylabel('Precipitation (in)', c= 'k')
plt.legend()
plt.savefig('Monthly and Yearly Total Precipitation.png')
```

```python
monthly_precip2 = weather.groupby(per)['totalprecipIn'].agg('mean')
yearly_precip2 = weather.groupby(per2)['totalprecipIn'].agg('mean')
plt.style.use('seaborn')
plt.title('Monthly and Yearly Average Precipitation')
monthly_precip2.plot(c='y', label = 'Monthly Average')
yearly_precip2.plot(c='k', label = 'Yearly Average')
plt.xlim('2009','2020')
plt.ylim(0 , 1)
plt.xlabel('Year')
plt.ylabel('Precipitation (in)')
plt.legend()
plt.savefig('Avg Total Precip.png')
```

```python
monthly_precip2 = weather.groupby(per)['totalprecipIn'].agg('mean')[5:139]
yearly_precip2 = weather.groupby(per2)['totalprecipIn'].agg('mean')[1:13]
monthly_precip2.plot(c='m', label = 'monthly std dev')
yearly_precip2.plot(c='k', label = 'yearly std dev')
plt.xlabel('Year')
plt.legend()
plt.ylim(0, 1)
plt.ylabel('Precipitation (in)')
plt.title('Standard Deviation of Monthly and Yearly Precipitation')
plt.savefig('Std Dev Total Precip.png')
```
